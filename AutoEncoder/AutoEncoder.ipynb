```python
import numpy as np
from sklearn import datasets 
```

```python
iris = datasets.load_iris() 
data = iris['data'] 
print(data)
```

```python
(data).shape
```

```python
from sklearn import preprocessing #归一化处理特别重要
new_data = preprocessing.normalize(data).T
new_data.shape
```

```python
#dims：4-5-5-4
def init(dims):
    np.random.seed(3)
    parameters = {}
    n = len(dims)
    for i in range(1,n):
        parameters['W'+str(i)] = np.random.randn(dims[i],dims[i-1])
        parameters['b'+str(i)] = np.zeros([dims[i],1])
    return parameters
```

```python
parameters = init([4,5,5,4])
print(parameters)
```

```python
print("W1 : ",parameters["W1"])
print("b1 : ",parameters["b1"])
print("W2 : ",parameters["W2"])
print("b2 : ",parameters["b2"])
print("W3 : ",parameters["W3"])
print("b3 : ",parameters["b3"])
```

```python
def forward_propagation(data,parameters):
    n = len(parameters)//2
    ZA = {}
    a = data
    for i in range(0,n):
        z = np.dot(parameters['W'+str(i+1)],a)+parameters['b'+str(i+1)]
        ZA['z'+str(i+1)] = a
        a = np.tanh(z)
        ZA['a'+str(i+1)] = a
    return a,ZA
```

```python
np.random.seed(4)
dataSet = np.random.randn(10,4)
a,ZA = forward_propagation(dataSet.T,parameters)
print(a)
print(ZA)
```

```python
def lossFunction(dataY,a,parameters,lamda=0.005):
    m = dataY.shape[1]
    #print(m)
    n = len(parameters)//2
    W = 0.0
    for i in range(0,n):
        W += (parameters['W'+str(i+1)]**2).sum()
    J = (np.power((dataY-a),2).sum()/m + lamda * W)/2 
    #J = (np.power((dataY-a),2).sum()/m)/2 
    return J
```

```python
print(dataSet.T.shape)
print(a.shape)
J = lossFunction(dataSet.T,a,parameters)
print(J)
```

```python
def back_propation(data,dataY,a,ZA,parameters,alpha=0.05): #在autoEncoder中data和dataY一样 alpha:学习速率
    m = dataY.shape[1]
    dparams = {}
    n = len(parameters)//2
    dZn = a - dataY
    dWn = np.dot(dZn,ZA['a'+str(n-1)].T)/m
    dbn = np.sum(dZn,axis = 1, keepdims = True)/m
    dparams['dZ'+str(n)] = dZn
    dparams['dW'+str(n)] = dWn
    dparams['db'+str(n)] = dbn
    parameters['W'+str(n)] = parameters['W'+str(n)] - alpha*dWn
    parameters['b'+str(n)] = parameters['b'+str(n)] - alpha*dbn
    for i in reversed(range(n-1)): #1,0
        dZi = np.multiply(np.dot(parameters["W"+str(i+1+1)].T,dparams["dZ"+str(i+1+1)]),(1 - np.power(ZA['a'+str(i+1)], 2)))
        if i != 0:
            dWi = np.dot(dZi,ZA['a'+str(i)].T)/m
        else:
            dWi = np.dot(dZi,data.T)/m
        dbi = np.sum(dZi,axis = 1, keepdims = True)/m
        dparams['dZ'+str(i+1)] = dZi
        dparams['dW'+str(i+1)] = dWi
        dparams['db'+str(i+1)] = dbi
        parameters['W'+str(i+1)] = parameters['W'+str(i+1)] - alpha*dWi
        parameters['b'+str(i+1)] = parameters['b'+str(i+1)] - alpha*dbi
    return parameters,dparams
```

```python
parameters,grads = back_propation(dataSet.T,dataSet.T,a,ZA, parameters)
print ("dW1 = "+ str(grads["dW1"]))
print ("db1 = "+ str(grads["db1"]))
print ("dW2 = "+ str(grads["dW2"]))
print ("db2 = "+ str(grads["db2"]))
print ("dW3 = "+ str(grads["dW3"]))
print ("db3 = "+ str(grads["db3"]))
```

```python
def autoEncoder(X,Y,dims,num_iterations = 10000):
    parameters = init(dims)
    for i in range(0,num_iterations):
        A,cache = forward_propagation(X, parameters)
        parameters,grads = back_propation(X,Y,A,cache,parameters)
        if i%1000 == 0:
            cost = lossFunction(Y,A,parameters)
            print("迭代：%d次，损失函数为：%f"%(i,cost))
    return parameters
```

```python
best_params = autoEncoder(new_data,new_data,[4,5,5,4])
```

```python
#print(best_params)
print("W1 = " + str(best_params["W1"]))
print("b1 = " + str(best_params["b1"]))
print("W2 = " + str(best_params["W2"]))
print("b2 = " + str(best_params["b2"]))
print("W3 = " + str(best_params["W3"]))
print("b3 = " + str(best_params["b3"]))
```

```python
def predict(X,parameters):
    A ,cache = forward_propagation(X, parameters)
    predictions = A
    return predictions
```

```python
predictData = predict(new_data,best_params)
```
