```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\miaohualin\\Desktop\\Python_Project\\Data_Mining\\scmd_CalculationData.csv")
df
```

```python
plt.figure(figsize=(10,9))
plt.scatter(df["winddirection"], df["windspeed1"], c = "red")
plt.rc('font', family='STXihei', size=15) 
plt.title("风速与风向的函数")
plt.xlabel("风向（。）")
plt.ylabel("风速（km/h）")
plt.grid(color='#95a5a6',linestyle='--',linewidth=3,axis='both',alpha= 0.4) 
fig = plt.gcf()
```

```python
fig.savefig("C:\\Users\\miaohualin\\Desktop\\Python_Project\\Data_Mining\\1风速与风向的函数.png")
```

```python
plt.figure(figsize=(10,9))
plt.scatter(df["windspeed1"],df["power_max"] , c = "red")
plt.scatter(df["windspeed1"],df["power_min"] , c = "yellow")
plt.scatter(df["windspeed1"],df["power"] , c = "blue")
plt.scatter(df["windspeed1"],df["power_dev"] , c = "green")
plt.legend(loc='upper left')
plt.rc('font', family='STXihei', size=15) 
plt.title("功率特效散点图")
plt.xlabel("风速（km/h）")
plt.ylabel("功率（kw）")
plt.grid(color='#95a5a6',linestyle='--',linewidth=3,axis='both',alpha= 0.4) 
fig = plt.gcf()
```
